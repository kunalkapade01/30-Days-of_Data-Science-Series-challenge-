```python
# Import library:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# ML library:
from sklearn.datasets import load_iris
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
# Load iris datasets:
iris = load_iris()
x = iris.data
y = iris.target
```

```python
#create dataframe :
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
# Spliting the data into train and test:
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=12)
```

```python
# Train the LDA model:
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
```

```python
# Making prediction :
y_pred = lda.predict(x_test)
y_pred
```

```python
# Evaluating  the model :
Accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
```

```python
# print the model :
print(f"ùêÄùêúùêúùêÆùê´ùêöùêúùê≤: {Accuracy}")
print(f"ùêÇùê®ùêßùêüùêÆùê¨ùê¢ùê®ùêß ùêåùêöùê≠ùê´ùê¢ùê±: \n{conf_matrix}")
print(f"ùêÇùê•ùêöùê¨ùê¨ùê¢ùêüùê¢ùêúùêöùê≠ùê¢ùê®ùêß ùêëùêûùê©ùê®ùê´ùê≠: \n{class_report}")
```

```python
#Transfer the data for visualization:
x_lda = lda.transform(x)
x_lda
```

```python
# plotting the LDA result:
plt.figure(figsize=(8, 6))
sns.scatterplot(x=x_lda[:, 0], y=x_lda[:, 1], hue=iris.target_names[y], palette="Set1")
plt.xlabel("LDA component 1")
plt.ylabel("LDA componenet 2")
plt.title("LDA of Iris Dataset")
plt.show()
```

- ùêÇùê®ùêßùêúùêûùê©ùê≠:
        Linear Discriminant Analysis (LDA) is a classification and dimensionality reduction technique that aims to project data points onto a lower-dimensional space while maximizing the separation between multiple classes. It achieves this by finding the linear combinations of features that best separate the classes. LDA assumes that the different classes generate data based on Gaussian distributions with the same covariance matrix.

ùêäùêûùê≤ ùêíùê≠ùêûùê©ùê¨ :
    
1. ùêÇùê®ùê¶ùê©ùêÆùê≠ùêû ùê≠ùê°ùêû ùêåùêûùêöùêß ùêïùêûùêúùê≠ùê®ùê´ùê¨: Compute the mean vector for each class.
2. ùêÇùê®ùê¶ùê©ùêÆùê≠ùêû ùê≠ùê°ùêû ùêíùêúùêöùê≠ùê≠ùêûùê´ ùêåùêöùê≠ùê´ùê¢ùêúùêûùê¨:
    - ùêñùê¢ùê≠ùê°ùê¢ùêß-ùêÇùê•ùêöùê¨ùê¨ ùêíùêúùêöùê≠ùê≠ùêûùê´ ùêåùêöùê≠ùê´ùê¢ùê±: Measures the scatter (spread) of features within each class.
    - ùêÅùêûùê≠ùê∞ùêûùêûùêß-ùêÇùê•ùêöùê¨ùê¨ ùêíùêúùêöùê≠ùê≠ùêûùê´ ùêåùêöùê≠ùê´ùê¢ùê±: Measures the scatter of the means of each class.
3. ùêíùê®ùê•ùêØùêû ùê≠ùê°ùêû ùêÜùêûùêßùêûùê´ùêöùê•ùê¢ùê≥ùêûùêù ùêÑùê¢ùê†ùêûùêßùêØùêöùê•ùêÆùêû ùêèùê´ùê®ùêõùê•ùêûùê¶: Compute the eigenvalues and eigenvectors for the scatter matrices to find the linear discriminants.
4. ùêíùê®ùê´ùê≠ ùêöùêßùêù ùêíùêûùê•ùêûùêúùê≠ ùêãùê¢ùêßùêûùêöùê´ ùêÉùê¢ùê¨ùêúùê´ùê¢ùê¶ùê¢ùêßùêöùêßùê≠ùê¨: Sort the eigenvalues in descending order and select the top eigenvectors to form a matrix of linear discriminants.
5. ùêèùê´ùê®ùê£ùêûùêúùê≠ ùê≠ùê°ùêû ùêÉùêöùê≠ùêö: Transform the original data onto the new subspace using the matrix of linear discriminants.
