```python
# Import library :
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_digits
from sklearn.metrics import accuracy_score
from scipy.stats import randint
```

```python
# Load dataseta :
digits = load_digits()
x, y = digits.data, digits.target
```

```python
x
```

```python
y
```

```python
# Define the model & hyperparameter search space :
model = RandomForestClassifier()
param_dist = {
    "n_estimators" : randint(10, 200),
    "max_depth" : randint(5, 50),
    "min_samples_split" : randint(2, 20),
    "min_samples_leaf" : randint(1, 20),
    "max_features" : ["sqrt", "log2", None]
}
```

```python
# Randomized search with cross-validation :
random_search = RandomizedSearchCV(model, param_distributions=param_dist, n_iter=100, cv=5, scoring='accuracy', verbose=1, n_jobs=-1)
random_search.fit(x, y)
```

```python
# print hyperparameters and score :
print("ğğğ¬ğ­ ğ‡ğ²ğ©ğğ«ğ©ğšğ«ğšğ¦ğğ­ğğ«ğ¬ ğ…ğ¨ğ®ğ§ğ: ")
print(random_search.best_params_)
```

```python
print("ğğğ¬ğ­ ğ€ğœğœğ®ğšğ«ğœğ² ğ’ğœğ¨ğ«ğ ğ…ğ¨ğ®ğ§ğ: ")
print(random_search.best_score_)
```
