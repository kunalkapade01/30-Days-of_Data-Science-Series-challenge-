# k-Means Clustering:

ùêä-ùêåùêûùêöùêßùê¨ ùê¢ùê¨ ùêöùêß ùêÆùêßùê¨ùêÆùê©ùêûùê´ùêØùê¢ùê¨ùêûùêù ùê•ùêûùêöùê´ùêßùê¢ùêßùê† ùêöùê•ùê†ùê®ùê´ùê¢ùê≠ùê°ùê¶ ùêÆùê¨ùêûùêù ùêüùê®ùê´ ùêúùê•ùêÆùê¨ùê≠ùêûùê´ùê¢ùêßùê† ùê≠ùêöùê¨ùê§ùê¨..

```python
# Import liubrary:
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Example data:
np.random.seed(0)
x = np.vstack((np.random.normal(0, 1, (100, 2)),
              np.random.normal(5, 1, (100, 2)),
              np.random.normal(-5, 1, (100, 2))))
```

```python
# Applying k-Means clustering:
k = 3
kmeans = KMeans(n_clusters=k, random_state=12)
y_kmeans = kmeans.fit_predict(x)
```

```python
# Plotting the Clusters:
plt.figure(figsize=(8,6))
sns.scatterplot(x=x[:, 0], y=x[:, 1], hue=y_kmeans, palette="viridis",
               s=50, edgecolor="k")
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
           s=300, c="red", label="Centroids")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("k-Means Clustering")
plt.legend()
plt.show()
```

#  ùêÑùê•ùêõùê®ùê∞ ùêåùêûùê≠ùê°ùê®ùêù ùêÑùê±ùêöùê¶ùê©ùê•ùêû:

ùêÑùê•ùêõùê®ùê∞ ùêåùêûùê≠ùê°ùê®ùêù ùê≠ùê® ùêüùê¢ùêßùêù ùê≠ùê°ùêû ùê®ùê©ùê≠ùê¢ùê¶ùêöùê• ùêßùêÆùê¶ùêõùêûùê´ ùê®ùêü ùêúùê•ùêÆùê¨ùê≠ùêûùê´ùê¨ :

```python
wcss = []   #Within-Cluster Sum of Squares.
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=12)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.figure(figsize=(8,6))
plt.plot(range(1, 11), wcss, marker="*")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

ùêÇùê®ùêßùêúùêûùê©ùê≠ :
         k-Means is an ùêîùêßùê¨ùêÆùê©ùêûùê´ùêØùê¢ùê¨ùêûùêù ùêãùêûùêöùê´ùêßùê¢ùêßùê† ùêÄùê•ùê†ùê®ùê´ùê¢ùê≠ùê°ùê¶ used for clustering tasks. The goal is to partition a dataset into \( k \) clusters, where each data point belongs to the cluster with the nearest mean. It is an iterative algorithm that aims to minimize the variance within each cluster.

ùêìùê°ùêû ùê¨ùê≠ùêûùê©ùê¨ ùê¢ùêßùêØùê®ùê•ùêØùêûùêù ùê¢ùêß ùê§-ùêåùêûùêöùêßùê¨ ùêúùê•ùêÆùê¨ùê≠ùêûùê´ùê¢ùêßùê† ùêöùê´ùêû:
    
- ùêàùêßùê¢ùê≠ùê¢ùêöùê•ùê¢ùê≥ùêöùê≠ùê¢ùê®ùêß: Choose \( k \) initial cluster centroids randomly.
- ùêÄùê¨ùê¨ùê¢ùê†ùêßùê¶ùêûùêßùê≠: Assign each data point to the nearest cluster centroid.
- ùêîùê©ùêùùêöùê≠ùêû: Recalculate the centroids as the mean of all points in each cluster.
- ùêëùêûùê©ùêûùêöùê≠: Repeat steps 2 and 3 until the centroids do not change significantly or a maximum number of iterations is reached.

 ùêÑùêØùêöùê•ùêÆùêöùê≠ùê¢ùê®ùêß ùêåùêûùê≠ùê´ùê¢ùêúùê¨: ùêñùêÇùêíùêí

- Within-Cluster Sum of Squares (WCSS): Measures the compactness of the clusters. Lower WCSS indicates more compact clusters.
- Silhouette Score: Measures the separation between clusters. Values range from -1 to 1, with higher values indicating better-defined clusters.
