```python
# Import library :
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
# Example time series data :
np.random.seed(42)
data_range = pd.date_range(start="1/1/2020", periods=365)
data = pd.Series(np.random.randn(len(data_range)), index=data_range)
```

```python
data
```

```python
type(data)
```

```python
# plotting the time series data :
plt.figure(figsize=(10, 6))
plt.plot(data)
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series data")
plt.grid(True)
plt.show()
```

```python
# Fit ARIMA model :
model = ARIMA(data, order=(1, 1, 1))  # Example order, replace with appropriate values
model_fit = model.fit()
```

```python
# Forecasting future vakues :
forecast_steps = 30   # Number of steps ahead to forecast.
forecast = model_fit.forecast(steps=forecast_steps)
```

```python
# Plotting the forecast :
plt.figure(figsize=(10, 6))
plt.plot(data, label="observed")
plt.plot(forecast, label="Forecast", linestyle="--", c="red")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("ARIMA Forecasting")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Evalute forecast accuracy (example with RMSE) :
test_data = pd.Series(np.random.randn(forecast_steps))  # Example test data, replace with actual test data.
```

```python
# print RMSE :
rmse = np.sqrt(mean_squared_error(test_data, forecast))
```
