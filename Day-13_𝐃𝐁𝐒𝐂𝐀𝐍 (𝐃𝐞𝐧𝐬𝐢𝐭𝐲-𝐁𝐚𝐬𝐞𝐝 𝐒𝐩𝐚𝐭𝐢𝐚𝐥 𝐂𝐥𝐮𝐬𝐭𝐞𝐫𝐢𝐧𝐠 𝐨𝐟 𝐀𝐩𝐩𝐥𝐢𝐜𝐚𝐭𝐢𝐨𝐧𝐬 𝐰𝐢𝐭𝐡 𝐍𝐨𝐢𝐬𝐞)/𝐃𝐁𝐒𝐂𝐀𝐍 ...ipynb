```python
# Import Library :
import pandas as pd
import numpy as np
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Generate datasets of (make_moons datasets):
X, y = make_moons(n_samples=300, noise=0.1, random_state=12)
```

```python
X
```

```python
y
```

```python
# Apply DBSCAN :
epsilon = 0.2
min_samples = 5
db = DBSCAN(eps=epsilon, min_samples=min_samples)
clusters = db.fit_predict(X)
```

```python
# Adding cluster labels to the dataframe :
df = pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Cluster"] = clusters
```

```python
# plottig the clusters :
plt.figure(figsize=(8, 6))
sns.scatterplot(x="Feature 1", y="Feature 2", hue="Cluster", palette="Set1", data=df)
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("DBSCAN Clustering")
plt.show()
```

ùêÇùê®ùêßùêúùêûùê©ùê≠ :
          DBSCAN is an unsupervised clustering algorithm that groups together points that are closely packed, and marks points that are in low-density regions as outliers. It is particularly effective for identifying clusters of arbitrary shape and handling noise in the data.

 Key Parameters :
        
- ùêÑùê©ùê¨ùê¢ùê•ùê®ùêß (Œµ): The maximum distance between two points to be considered neighbors.
- ùêåùê¢ùêßùêèùê≠ùê¨ : The minimum number of points required to form a dense region (a cluster).


ùêäùêûùê≤ ùêìùêûùê´ùê¶ùê¨ :
    
- ùêÇùê®ùê´ùêû ùêèùê®ùê¢ùêßùê≠: A point with at least MinPts neighbors within a radius of Œµ.
- ùêÅùê®ùê´ùêùùêûùê´ ùêèùê®ùê¢ùêßùê≠: A point that is not a core point but is within the neighborhood of a core point.
- ùêçùê®ùê¢ùê¨ùêû ùêèùê®ùê¢ùêßùê≠: A point that is neither a core point nor a border point (outlier).
