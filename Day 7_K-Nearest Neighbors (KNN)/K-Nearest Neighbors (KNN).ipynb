# K-Nearest Neighbors (KNN):

```python
# Import necessary library:
import pandas as pd
import numpy as np
import pickle
```

```python
# ML library:
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# Example data (Iris Datasets):
from sklearn.datasets import load_iris
iris = load_iris()    # load iris datsets:
```

```python
# Create Dataframe:
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
# Split the iris datasets into x and y:
x = iris.data[:, :2]   #Using sepal length and sepal width as feature.
x
```

```python
y = iris.target    # as target column:
y
```

```python
# Spliting the data into trainig and testing sets:
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25,
                                                   random_state=12)
```

```python
# Creating & training the KNN model with k=5:
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)
```

```python
# Save the trained model to a file
with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
```

```python
print("Modle trained and saved..")
```

```python
# Making prediction:
y_pred = model.predict(X_test)
y_pred
```

```python
# Evaluation the model and print:
accuarcy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuarcy}")
```

```python
# Print the evaluation:
conf_matrix = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix: \n{conf_matrix}")
```

```python
class_report = classification_report(y_test, y_pred)
print(f"Classification Report: \n{class_report}")
```

```python
# Plotting the decision boundary:
def plot_decision_boundary(x, y, model):
    h = 0.2   # Step size in the mesh.
    x_min, x_max = x[:, 0].min() - 1, x[:, 0].max() + 1
    y_min, y_max = x[:, 1].min() - 1, x[:, 1].max() + 1
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    z = z.reshape(xx.shape)
    plt.contourf(xx, yy, z, alpha=0.8)
    
    sns.scatterplot(x=x[:, 0], y=x[:, 1], hue=y, palette="bright",
                    edgecolor="k", s=50)
    plt.xlabel("Sepal Length")
    plt.ylabel("Petal Length")
    plt.title("KNN Decision Boundary")
    plt.show()
```

```python
# Function call:
plot_decision_boundary(X_test, y_test, model)
```
