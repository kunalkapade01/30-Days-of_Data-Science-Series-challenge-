# Principal Component Analysis (PCA):-

```python
#Import linbray:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#Import ML library:
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Example data (Iris) datasets:
from sklearn.datasets import load_iris
iris = load_iris()
iris
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df
```

```python
x = iris.data
y = iris.target
```

```python
x
```

```python
y
```

```python
# Standardizing the features:
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
#Applying PCA model:
pca = PCA(n_components=2)
x_pca = pca.fit_transform(x_scaled)
```

```python
# Plotting the principal components;
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:, 0], x_pca[:, 1], c=y, cmap="viridis", edgecolors="k", s=50)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA of Iris Datasets")
plt.colorbar()
plt.show()
```

```python
#Explained variance:
exp_variance = pca.explained_variance_ratio_
print(f"ð„ð±ð©ð¥ðšð¢ð§ðžð ð•ðšð«ð¢ðšð§ðœðž ð›ð² ð‚ð¨ð¦ð©ð¨ð§ðšð§ð­ ðŸ: {exp_variance[0]:2f}")
print(f"ð„ð±ð©ð¥ðšð¢ð§ðžð ð•ðšð«ð¢ðšð§ðœðž ð›ð² ð‚ð¨ð¦ð©ð¨ð§ðšð§ð­ ðŸ: {exp_variance[1]:2f}")
```

# Summary :-

ðð‚ð€ ð‚ð¨ð§ðœðžð©ð­ : Principal Component Analysis (PCA) is a dimensionality reduction technique used to transform a large set of correlated features into a smaller set of uncorrelated features called principal components. These principal components capture the maximum variance in the data while reducing the dimensionality.

ð“ð¡ðž ð¬ð­ðžð©ð¬ ð¢ð§ð¯ð¨ð¥ð¯ðžð ð¢ð§ ðð‚ð€ ðšð«ðž:
- Standardization: Normalize the data to have zero mean and unit variance.
- Covariance Matrix Computation: Compute the covariance matrix of the features.
- Eigenvalue and Eigenvector Decomposition: Compute the eigenvalues and eigenvectors of the covariance matrix.
- Principal Components Selection: Select the top \(k\) eigenvectors corresponding to the largest eigenvalues to form the principal components.
- Transformation: Project the original data onto the new subspace formed by the selected principal components.

ððžð§ðžðŸð¢ð­ð¬ ð¨ðŸ ðð‚ð€:
- ð‘ðžðð®ðœðžð¬ ðƒð¢ð¦ðžð§ð¬ð¢ð¨ð§ðšð¥ð¢ð­ð²: Simplifies the dataset by reducing the number of features.
- ðˆð¦ð©ð«ð¨ð¯ðžð¬ ððžð«ðŸð¨ð«ð¦ðšð§ðœðž: Speeds up machine learning algorithms and reduces the risk of overfitting.
- ð”ð§ðœð¨ð¯ðžð«ð¬ ð‡ð¢ðððžð§ ððšð­ð­ðžð«ð§ð¬: Helps visualize the underlying structure of the data.
